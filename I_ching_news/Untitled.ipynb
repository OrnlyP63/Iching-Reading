```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

```python
def is_not_digit(words):
    for s in words:
        if s.isdigit():
            return False
        return True
        
```

```python
words = [w for w in words if not w in stop_words and is_not_digit(w)]
```
